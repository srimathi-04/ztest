```python
import pandas as pd
from statsmodels.stats import weightstats as stests
```

```python
df=pd.read_csv("/content/Iris.csv")
df
```

```python
df.head()
```

```python
df[['SepalWidthCm','PetalWidthCm']].describe()
```

```python
ztest ,pvalue = stests.ztest(df['SepalWidthCm'], x2=None, value=150)
print(float(pvalue))
print(ztest)
```

```python
if pvalue<0.05:
    print("Null hypothesis rejected , Alternative hypothesis accepted")
else:
    print("Null hypothesis accepted , Alternative hypothesis rejected")
```

```python
ztest,pval1=stests.ztest(df['PetalLengthCm'],df['SepalLengthCm'],value=125)
print(float(pval1))
```

```python
if pval1>0.05:
    print("Null hypothesis rejected , Alternative hypothesis accepted")
else:
    print("Null hypothesis accepted , Alternative hypothesis rejected")
```
